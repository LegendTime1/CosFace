```python
import shutil

# Path to the ZIP file
zip_file_path = '/content/CosFace/dataset/CASIA-maxpy-clean_fix.zip'  # Adjust the path if needed

# Directory to extract the contents
extract_dir = '/content/CosFace/dataset'  # Directory to extract to

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Extract the contents
with open(zip_file_path, 'rb') as f_in:
    with open(os.path.join(extract_dir, 'temp.zip'), 'wb') as f_out:
        shutil.copyfileobj(f_in, f_out)

# Rename the temp file to .zip
os.rename(os.path.join(extract_dir, 'temp.zip'), os.path.join(extract_dir, 'CASIA-maxpy-clean_fix.zip'))

# Unzip the ZIP file
with zipfile.ZipFile(os.path.join(extract_dir, 'CASIA-maxpy-clean_fix.zip'), 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# List the extracted files
extracted_files = os.listdir(extract_dir)
print('Files extracted:', extracted_files)
```
